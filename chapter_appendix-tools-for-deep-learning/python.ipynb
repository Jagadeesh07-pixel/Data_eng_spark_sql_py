```python
#reverse string
def rev_str(s):
  res =''
  for i in s:
    res = i+res
  return res
print(rev_str('jagadeesh'))
```

```python
#find dup & unique
def unq(s):
  unique = []
  dup = []
  for i in s:
    if i in unique:
      dup.append(i)
    else:
      unique.append(i)
  return unique, dup
print(unq('jagadeesh'))
```

```python
#count of words
s = 'prepare well for interview and clear the interview'
words = s.split()
freq ={}
for word in words:
  if word in freq:
    freq[word] += 1
  else:
    freq[word] = 1
print(freq)
#without split
count = 1
for word in s:
  if word == ' ':
    count+=1
print(count)
```

```python
#remove spcl chars
s = 'hello@123'
res = ''
for ch in s:
  if ch.isalnum() or ch == ' ':
    res += ch
print(res)

x = 'whdgyu$%^2134%^&'
a = ['!','@','#','$','%','^','&','*']
res1 = ''
for ch in x:
  if ch not in a:
    res1 += ch
print(res1)
```

```python
#most common element
lst =[1,2,3,4,5,2,3,2,2,2,13,4]
from collections import Counter
counter = Counter(lst)
print(counter)
most_common_element, count = counter.most_common(1)[0]
print(str(most_common_element)+':'+str(count)+'times')
#first non repeating value
for i in lst:
  if counter[i] == 1:
    break
print(i)
```

```python
#remove duplicate char
res = "".join(dict.fromkeys('string'))
print(res)
sort1 = [1,2,3,2,3]
print(set(sort1))
#check two strings are anagrams
s1 = 'silent'
s2 = 'listen'
if sorted(s1)==sorted(s2):
  print('anagram')
else:
  print('not anagram')

#count vowels and constants
vowels = ['a','e','i','o','u']
s ='johnsnow'
count_vowel = 0
count_const = 0
for i in s:
  if i in vowels:
    count_vowel += 1
  else:
    count_const += 1
print(count_vowel,count_const)

#reverse words in sentence
s = 'all the best'
res = ''
words = s.split()
for word in words:
  res = word+' '+res
print(res)
#palindrome
s = 'amma'
res = ''
for i in s:
  res = i + res
if s == res:
  print('palindrome')
else:
  print('not a palindrome')

```
