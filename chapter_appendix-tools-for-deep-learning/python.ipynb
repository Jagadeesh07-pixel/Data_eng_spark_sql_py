```python
#reverse string
def rev_str(s):
  res =''
  for i in s:
    res = i+res
  return res
print(rev_str('jagadeesh'))
```

```python
#find dup & unique
def unq(s):
  unique = []
  dup = []
  for i in s:
    if i in unique:
      dup.append(i)
    else:
      unique.append(i)
  return unique, dup
print(unq('jagadeesh'))
```
